#### Стиль линии графика


Стиль линии графика задается через параметр linestyle, 
который может принимать значения из приведенной ниже таблицы.

Значение параметра | Описание
------------------|-------
$-$   или $solid$ | Непрерывная линия
$--$   или $dashed$ | Штриховая линия
$-.$  или $dashdot$ | Штрихпунктирная линия
$:$   или $dotted$ | Пунктирная линия
"$None$" или "пробел" | Не отображать линию

### Методы Axes и pyplot

Раньше мы создавали графики без всяких Figure и Axes.   

Дело в том что почти все методы axes присутствуют в модуле pyplot.   

Например, при вызове plt.title('График') модуль pyplot вызывает ax.set_title('График').   

Можно сказать, что модуль pyplot создает Figure и Axes автоматически.  

Фактически мы можем переписать весь наш пример следующим образом:   

```python
plt.plot([0, 1, 2, 3, 4], [0, 6, 7, 15, 19], linewidth = 3)
plt.scatter([0, 1, 2, 3, 4], [1, 3, 8, 12, 27], color = 'orange');
```

### Области Axes можно добавить двумя способами: add_subplot() и subplots()

#### Давайте рассмотрим случай add_subplot(): 

```python
#  Данные:
x = np.linspace(0, 10, 100)
y = np.sin(x)
img = y.reshape(5, 20)

#  Создаем "Figure" и "Axes":
fig = plt.figure()
ax_1 = fig.add_subplot(2, 1, 1)
ax_2 = fig.add_subplot(2, 1, 2)

#  Методы, отображающие данные:
ax_1.plot(x, y)
ax_2.imshow(img)

#  Добавление заголовков:
ax_1.set(title = 'Sin(x)')
ax_2.set(title = 'Изображение')
fig.tight_layout();
```

#### В случае subplots() все тоже самое, только к областям Axes мы обращаемся по индексу:

```python
#  Данные:
x = np.linspace(0, 10, 100)
y = np.sin(x)
img = y.reshape(5, 20)

#  Создаем "Figure" и "Axes":
fig, axes = plt.subplots(nrows = 2, ncols = 1)

#  Методы, отображающие данные:
axes[0].plot(x, y)
axes[1].imshow(img)

#  Добавление заголовков:
axes[0].set(title = 'Sin(x)')
axes[1].set(title = 'Изображение')
fig.tight_layout();
```

### plt.legend()

```python

# Строковое описание 	Код
# ‘best’     в зависимости от расположение графика
# ‘upper right’ 	1
# ‘upper left’ 	2
# ‘lower left’ 	3
# ‘lower right’ 	4
# ‘right’ 	    5
# ‘center left’ 	6
# ‘center right’ 	7
# ‘lower center’ 	8
# ‘upper center’ 	9
# ‘center’ 	    10
# ‘upper left’ 	0
```

```python
x = [1, 5, 10, 15, 20]
y1 = [1, 7, 3, 5, 11]
y2 = [4, 3, 1, 8, 12]
locs = ['best', 'upper right', 'upper left', 'lower left', 
        'lower right', 'right', 'center left', 'center right', 
        'lower center', 'upper center', 'center']
plt.figure(figsize=(12, 12))
for i in range(3):
    for j in range(4):
        if i * 4 + j < 11:
            plt.subplot(3, 4, i * 4 + j + 1)
            plt.title(locs[i * 4 + j])
            plt.plot(x, y1, 'o-r', label='line 1')
            plt.plot(x, y2, 'o-.g', label='line 2')
            plt.legend(loc=locs[i * 4 + j])
        else:
            break
```
