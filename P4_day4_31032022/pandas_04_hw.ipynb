```python
import pandas as pd
import numpy as np
```

```python
def haversine(lat1, lon1, lat2, lon2):
    RADIUS = 6371.2
    lat1, lon1, lat2, lon2 = map(np.deg2rad, [lat1, lon1, lat2, lon2])
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = np.sin(dlat / 2) ** 2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon / 2) ** 2
    c = 2 * np.arcsin(np.sqrt(a))
    total_km = RADIUS * c
    return total_km
```

```python
russia = pd.read_csv("cities.csv") 
russia.loc[:,'Население'] = pd.to_numeric(russia.loc[:,'Население'], errors='coerce', downcast='unsigned')
print(russia.iloc[0]['Регион'])
#russia.filter(like=russia.iloc[0]['Регион']), axis=0)
value = russia.loc[((russia['Регион'] == russia.iloc[0]['Регион']) & (russia['Признак центра района или региона'] == 2)), 'Широта'].item()
value
```

```python
russia_cfo = russia.query('`Федеральный округ` == "Центральный"')
russia_cfo.loc[:,'Расстояние'] = haversine(russia.query('Регион == "Москва"').loc[:,'Широта'].item(), russia.query('Регион == "Москва"').loc[:,'Долгота'].item(), russia_cfo.loc[:,'Широта'].values, russia_cfo.loc[:,'Долгота'].values).round(2)
hw1 = russia_cfo.query('Расстояние <= 100').query('Население <= 50000')
hw1
```

```python
russia_sfo = russia.query('`Федеральный округ` == "Сибирский"')
russia_sfo.loc[:,'Расстояние'] = haversine(float(russia_sfo.query('Город == "Новосибирск"')['Широта']), float(russia_sfo.query('Город == "Новосибирск"')['Долгота']), russia_sfo.loc[:,'Широта'].values, russia_sfo.loc[:,'Долгота'].values).round(2)
hw2 = russia_sfo.query('150 <= Расстояние <= 200').query('100000 <= Население <= 200000')
hw2
```

```python
russia_dfo = russia.query('`Федеральный округ` == "Дальневосточный"')
russia_dfo.loc[:,'Расстояние'] = haversine(float(russia_dfo.query('Город == "Владивосток"')['Широта']), float(russia_dfo.query('Город == "Владивосток"')['Долгота']), russia_dfo.loc[:,'Широта'].values, russia_dfo.loc[:,'Долгота'].values).round(2)
hw3 = russia_dfo.query('10000 <= Население <= 20000').sort_values('Расстояние').head(5)
hw3
```

```python
def haversine_looping(df):
    distance_list = []
    for i in range(len(df)):
        d = haversine(df.loc[((df['Регион'] == df.iloc[i]['Регион']) & (df['Признак центра района или региона'] == 2)), 'Широта'].item(), df.loc[((df['Регион'] == df.iloc[i]['Регион']) & (df['Признак центра района или региона'] == 2)), 'Долгота'].item(), df.iloc[i]['Широта'], df.iloc[i]['Долгота'])
        distance_list.append(d)
    return distance_list

russia['Расстояние'] = haversine_looping(russia)


```
